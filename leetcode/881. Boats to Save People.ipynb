```cpp
#include <vector>
#include <bits/stdc++.h>
#include <algorithm>

using namespace std;

class Solution
{
public:
    vector<int> used_index = {};
    int numRescueBoats(vector<int> &people, int limit)
    {
        vector<int> arr = people;
        sort(arr.begin(), arr.end(), greater<int>());

        int boats = 0;
        auto is_used = [=](int num) -> bool {
            if (find(this->used_index.begin(), this->used_index.end(), num) != this->used_index.end())
            {
                return true;
            }
            else
            {
                return false;
            }
        };

        int target = 0;
        int length = arr.size();
        for (int i = 0; i < length; ++i)
        {
            int e = arr[i];

            if (is_used(i))
            {
                continue;
            }

            if (e <= limit)
            {
                this->used_index.push_back(i);
                target = limit - e;

                for (int j = i + 1; j < length; ++j)
                {
                    if (is_used(j))
                    {
                        continue;
                    }

                    e = arr[j];

                    if (e <= target)
                    {
                        this->used_index.push_back(j);
                        break;
                    }
                }

                boats += 1;
            }
        }

        return boats;
    }
};

int main(int argc, char *argv[])
{
    auto s = Solution();
    vector<int> people = {3, 2, 2, 1};
    int limit = 3;
    int r = s.numRescueBoats(people, limit);
    cout << "result: " << r << endl;
    return 0;
}
```
