```cpp
#include <iostream>
#include <ctime>

using namespace std;

class Solution
{
public:
    time_t parseTimeString(string date)
    {
        const char *dateString = date.c_str();
        struct tm timeinfo;
        int year, month, day;
        if (sscanf(dateString, "%d-%d-%d", &year, &month, &day) != EOF)
        {
            time_t rawtime;
            time(&rawtime);
            timeinfo = *localtime(&rawtime);
            timeinfo.tm_year = year - 1900;
            timeinfo.tm_mon = month - 1;
            timeinfo.tm_mday = day;
        }
        return mktime(&timeinfo);
    }

    time_t parseTimeStringAndGetTheStartingDay(string date)
    {
        const char *dateString = date.c_str();
        struct tm timeinfo;
        int year, month, day;
        if (sscanf(dateString, "%d-%d-%d", &year, &month, &day) != EOF)
        {
            time_t rawtime;
            time(&rawtime);
            timeinfo = *localtime(&rawtime);
            timeinfo.tm_year = year - 1900;
            timeinfo.tm_mon = 0;
            timeinfo.tm_mday = 1;
            timeinfo.tm_hour = 0;
            timeinfo.tm_min = 0;
            timeinfo.tm_sec = 0;
        }
        return mktime(&timeinfo);
    }
    int dayOfYear(string date)
    {
        //https://www.cplusplus.com/reference/ctime/mktime/
        //https://www.cplusplus.com/reference/ctime/
        auto time1 = this->parseTimeString(date);
        auto time2 = this->parseTimeStringAndGetTheStartingDay(date);
        auto seconds = difftime(time1, time2);
        return seconds / 60 / 60 / 24 + 1;
    }
};
```

````c++
int main(int argc, char *argv[])
{
    auto s = Solution();
    auto r = s.dayOfYear("2019-02-10");
    cout << "result: " << r << endl;
    return 0;
}
````
