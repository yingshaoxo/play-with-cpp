```cpp
#include <iostream>
#include <vector>
#include <math.h>
#include <ctype.h>
#include <bits/stdc++.h>
#include <map>

using namespace std;

class Solution
{
public:
    vector<string> getWords(string S)
    {
        vector<string> tokens;
        stringstream stream(S);
        string intermediate;

        while (getline(stream, intermediate, ' '))
        {
            tokens.push_back(intermediate);
        }

        return tokens;
    }

    void updateCounter(map<string, int>& d, vector<string> l) {
        for (auto v : l) {
            if (d.find(v) == d.end()) {
                d[v] = 1;
            } else {
                d[v] += 1;
            }
        }
    }

    vector<string> uncommonFromSentences(string A, string B)
    {
        vector<string> a = getWords(A);
        vector<string> b = getWords(B);
        vector<string> r;
        map<string, int> counter;
        updateCounter(counter, a);
        updateCounter(counter, b);

        for (auto& it: counter) {
            if (it.second == 1) {
                r.push_back(it.first);
            }
        }

        return r;
    }
};
```
