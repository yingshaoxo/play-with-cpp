```cpp
#include <iostream>
#include <bits/stdc++.h>
#include <math.h>
#include <set>

using namespace std;

/*
struct TreeNode
{
    int val;
    TreeNode *left;
    TreeNode *right;
    TreeNode() : val(0), left(nullptr), right(nullptr) {}
    TreeNode(int x) : val(x), left(nullptr), right(nullptr) {}
    TreeNode(int x, TreeNode *left, TreeNode *right) : val(x), left(left), right(right) {}
};
*/

class Solution
{
public:
    set<int, less<int> > set1;
    int getMinimumDifference(TreeNode *root)
    {
        set1.clear();
        travel(root);
        std::vector<int> vector1(set1.size());
        std::copy(set1.begin(), set1.end(), vector1.begin());
        int minV = INT_MAX;
        for (size_t i=0; i < vector1.size(); ++i) {
            if (i != 0) {
                int v2 = vector1[i];
                int v1 = vector1[i-1];
                int diff = v2 - v1;
                if (diff < minV) {
                    minV = diff;
                }
            }
        }
        return minV;
    }
    void travel(TreeNode *root) {
        if (root == nullptr) {
            return;
        }
        set1.insert(root->val);
        travel(root->left);
        travel(root->right);
    }
};
```
