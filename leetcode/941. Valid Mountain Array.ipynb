```cpp
#include <vector>
#include <algorithm>
#include <bits/stdc++.h> 

using namespace std;

class Solution {
public:
    bool validMountainArray(vector<int>& arr) {
        //5:18
        if (arr.size()==1) {
            return false;
        }
        bool up = false;
        bool down = false;
        int count = 0;
        for (int i=0; i < arr.size(); i++) {
            if (i != 0) {
                auto previous = arr[i-1];
                auto current = arr[i];
                if (previous == current) {
                    return false;
                }
                if (up == false) {
                    if ( current > previous) {
                        up = true;
                    } else if (current < previous) {
                        return false;
                    }
                }
                if (up == true) {
                    if ( current < previous) {
                        down = true;
                        for (int new_i=i; new_i < arr.size(); new_i++) {
                            auto previous = arr[new_i-1];
                            auto current = arr[new_i];
                            if (previous == current) {
                                return false;
                            }
                            if (current > previous){
                                return false;
                            }
                        }
                        return true;
                    }
                }
            }
        }
        return false;
        //5:28
    }
};
```
