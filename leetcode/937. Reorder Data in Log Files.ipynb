```cpp
#include <vector>
#include <algorithm>
#include <string>
#include <iostream>
#include <tuple>

using namespace std;

bool sort_tuple_by_our_rules(const tuple<string, string> &a, const tuple<string, string> &b) {
    if ((get<1>(a) == get<1>(b))) {
        return get<0>(a) < get<0>(b);
    }
    if ((get<1>(a) < get<1>(b)) && (get<1>(b) < get<1>(a))) {
        return get<0>(a) < get<0>(b);
    }
    else {
        return get<1>(a) < get<1>(b);
    };
}

class Solution
{
public:
    vector<string> reorderLogFiles(vector<string> &logs) {
        //6:57
        vector<tuple<string, string>> digits_list;
        vector<tuple<string, string>> letters_list;
        for (const auto &log : logs) {
            string identifier = "";
            string content = "";
            for (int i = 0; i < log.size(); i++) {
                if (log[i] == ' ') {
                    identifier = log.substr(0, i);
                    content = log.substr(i, log.size() - i);
                    break;
                }
            }
            bool is_digit_log = true;
            for (auto c : content) {
                if (!(isdigit(c) || c == ' ')) {
                    is_digit_log = false;
                    break;
                }
            }
            if (is_digit_log) {
                digits_list.push_back(make_tuple(identifier, content));
            }
            else {
                letters_list.push_back(make_tuple(identifier, content));
            }
        }
        sort(letters_list.begin(), letters_list.end(), sort_tuple_by_our_rules);
        vector<string> results;
        for ( auto each_tuple: letters_list) {
            results.push_back(get<0>(each_tuple) + get<1>(each_tuple));
        }
        for ( auto each_tuple: digits_list) {
            results.push_back(get<0>(each_tuple) + get<1>(each_tuple));
        }
        return results;
        //7:27
    }
};
```
